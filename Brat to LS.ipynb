```python
import csv
import json
import os
```

```python
filenames = set()
for filename in os.listdir('data'):
    filenames.add(filename.split('.')[0])
print(filenames)
```

```python
ls_data = []
for filename in filenames:
    task = {}
    with open(os.path.join('data', filename + '.txt')) as f:
        text = f.read()
    task['data'] = {'text': text}
    task['predictions'] = [{
        'model_version': 'crowd',
        'score': 0.99,
        'result': []
    }]
    with open(os.path.join('data', filename + '.ann')) as f:
        for line in f:
            id_, info, mention = line.split('\t')
            t, start, end = info.split()
            if t != 'PROP_PER':
                continue
            task['predictions'][0]['result'].append(
                {
                    "id": id_,
                    "from_name": "label",
                    "to_name": "text",
                    "type": "labels",
                    "value": {
                      "start": start,
                      "end": end,
                      "score": 0.99,
                      "text": mention,
                      "labels": [
                        "PER"
                      ]
                    }
                }
            )
    ls_data.append(task)
```
